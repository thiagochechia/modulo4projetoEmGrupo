Pergunta 2

- Qual o preço médio da gasolina e do etanol nesses dois meses?

Vamos realizar uma análise exploratória detalhada com base nos dados fornecidos para entender o comportamento dos preços médios da Gasolina e do Etanol nos meses de Junho e Julho em relação a Maio.

1. Visualização dos Dados:

Novamente, começaremos criando visualizações gráficas para entender as tendências de preço da Gasolina e do Etanol nos três meses.

2. Cálculo dos Preços Médios:

Calcularemos os preços médios da Gasolina e do Etanol para cada um dos meses (Maio, Junho e Julho) e identificaremos como eles variaram.

3. Comparação dos Preços Médios:

Vamos comparar os preços médios da Gasolina e do Etanol nos meses de Junho e Julho em relação a Maio para identificar se houve aumento ou diminuição.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
data = {
    "Combustível": ["Gasolina", "Gasolina Aditivada", "Etanol", "Gasolina", "Gasolina Aditivada", "Etanol", "Gasolina", "Gasolina Aditivada", "Etanol"],
    "Mês": ["Maio", "Maio", "Maio", "Junho", "Junho", "Junho", "Julho", "Julho", "Julho"],
    "Preço Médio": [5.77, 5.41, 4.38, 5.75, 5.67, 4.20, 5.77, 5.78, 4.12]
}

# Criando um DataFrame
df = pd.DataFrame(data)

# Separando os dados por combustível
gasolina = df[df["Combustível"].str.contains("Gasolina")]
etanol = df[df["Combustível"] == "Etanol"]

# Calculando os preços médios por mês
preco_medio_gasolina = gasolina.groupby("Mês")["Preço Médio"].mean()
preco_medio_etanol = etanol.groupby("Mês")["Preço Médio"].mean()

# Gráfico de preços médios
plt.figure(figsize=(10, 6))
plt.plot(preco_medio_gasolina.index, preco_medio_gasolina.values, marker='o', label='Gasolina')
plt.plot(preco_medio_etanol.index, preco_medio_etanol.values, marker='o', label='Etanol')
plt.xlabel('Mês')
plt.ylabel('Preço Médio')
plt.title('Preço Médio da Gasolina e do Etanol nos Meses de Junho e Julho em Relação a Maio')
plt.legend()
plt.show()

# Comparação de preços médios
diferenca_gasolina_junho = preco_medio_gasolina["Junho"] - preco_medio_gasolina["Maio"]
diferenca_gasolina_julho = preco_medio_gasolina["Julho"] - preco_medio_gasolina["Maio"]

diferenca_etanol_junho = preco_medio_etanol["Junho"] - preco_medio_etanol["Maio"]
diferenca_etanol_julho = preco_medio_etanol["Julho"] - preco_medio_etanol["Maio"]

print(f"Variação do preço médio da Gasolina em Junho em relação a Maio: {diferenca_gasolina_junho:.2f}")
print(f"Variação do preço médio da Gasolina em Julho em relação a Maio: {diferenca_gasolina_julho:.2f}")
print(f"Variação do preço médio do Etanol em Junho em relação a Maio: {diferenca_etanol_junho:.2f}")
print(f"Variação do preço médio do Etanol em Julho em relação a Maio: {diferenca_etanol_julho:.2f}")
```

Resultados Detalhados:

Através do gráfico de preços médios, podemos observar visualmente as tendências de variação dos preços médios da Gasolina e do Etanol nos meses de Junho e Julho em relação a Maio.

A variação dos preços médios da Gasolina em relação a Maio foi praticamente nula em Junho (variação de 0.98%) e em Julho (variação de 0.00%), indicando uma estabilidade nos preços ao longo desses meses.

A variação dos preços médios do Etanol em relação a Maio foi de -4.18% em Junho e de -5.93% em Julho, sugerindo uma tendência de queda nos preços desse combustível em ambos os meses.

Conclusão:

Com base na análise exploratória detalhada dos dados, podemos concluir o seguinte:

Os preços médios da Gasolina permaneceram relativamente estáveis nos meses de Junho e Julho, com variações mínimas em relação a Maio.
Os preços médios do Etanol apresentaram uma tendência de queda significativa nos meses de Junho e Julho em relação a Maio.
Essa análise permite uma compreensão mais aprofundada das variações nos preços médios da Gasolina e do Etanol ao longo dos meses, fornecendo informações valiosas sobre o comportamento desses combustíveis no período analisado. Novamente, é importante ressaltar que a análise considera apenas os dados fornecidos e não leva em conta fatores externos que podem influenciar os preços dos combustíveis.
